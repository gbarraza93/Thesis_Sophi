```python
import pandas as pd
from datetime import datetime
```

# Leyendo la tabla de datos final generada con el código de Python

1. Leemos la tabla CSV generada con el código de Python y generamos un "Pandas Data frame". Para esto, debemos pasar el nombre del archivo a la función `pandas.read_csv(<nombre del archivo>, sep = <separador>)`.
2. La función `display()` muestra la tabla con un formato reconocido por el Jupyter notebook.

```python
# Comenzamos poniendo el nombre del archivo csv que queremos leer. Si no se encuentra en la misma carpeta que
# este notebook, entonces debes poner el path global del archivo
file = "table_with_data_FINAL_nueva.csv"

df = pd.read_csv(file, sep=';') # Recuerda, se necesita especificar cual es el separador de columnas

display(df)
```

Si te fijas, la tabla de arriba ahora tiene la columna "*DataTimestampLOCAL*". De nuevo, para poder hacer las cosas más fácil, convertiremos esa columna como los nuevos índices.

```python
# Convert DataTimestampLOCAL to date and time format and round to minutes
def to_datetime(dt):
    return datetime.strptime(dt, "%Y-%m-%d %H:%M:%S").replace(second=0)

df["DataTimestampLOCAL"] = df["DataTimestampLOCAL"].apply(to_datetime)
# Set DataTimestampLOCAL as index
df.set_index("DataTimestampLOCAL", inplace=True)
display(df)
```

## 1. Eliminando columnas de la tabla

Digamos que queremos eliminar la columna *automatic_fee* de la tabla de datos. Para esto, solo requieres utilizar la función `drop(<nombre de la columna>, axis=1)`. *axis=1* especifica que es columna y no fila.

```python
df.drop("cong_mess_1", axis=1)
```

Lo anterior únicamente elimina la columna temporalmente. Para conservar el resultado, es necesario que lo asignes a una nueva variable o, en su defecto, volver a escribir sobre la misma variable (preferible).

```python
df = df.drop("cong_mess_1", axis=1) # ahora, la nueva tabla de datos no contiene la columna "automatic_fee"
display(df.sample(5)) # display(df.sample(<int>)) imprime una muestra aleatoria del número que le pases (esto solo es para imprimir menos lineas y ver los resultados mejor) 
```

Por el contrario, si deseamos conservar unas cuantas columnas y eliminar las restantes, solo es necesario proporcionar el nombre de las columnas que queremos conservar como argumento de la tabla. Aquí un ejemplo:

```python
df[["date", "time", "month_name", "day_name", "speed"]].head() # head() únicamente muestra las primeras ~ 5 filas
```

## 2. Insertando nuevas columnas a la tabla de datos

En tu caso particular, te gustaría agregar los nombres del día y del mes, basado en la fecha correspondiente en la columna *date*. Para esto, hacemos lo siguiente:

1. Para cada fecha en la columna *date*, extraemos el nombre del mes con la función de Pandas DateTimeIndex`month_name()`. Esta dunción se aplica a los índices de la tabla con timestamp *DataTimestampLOCAL*. 
2. Hacemos lo anterior para los nombres de los días también. La única diferencia es que debemos aplicar ahora la función `day_name()`.
3. Por último, utilizamos la función `df.insert(<posición>, <nombre para la columna>, <datos>)` para insertar una nueva columna en la posición requerida.

```python
month_name = df.index.month_name()
day_name = df.index.day_name()

# como ya tenemos las columnas month_name y day_name, las llamaremos distinto para que se vea la diferencia
df.insert(2, "month", month_name) # insertamos la columna en la posición 3 (a la derecha de "time")
df.insert(3, "day", day_name) # insertamos la columna en la posición 4 (a la derecha de "month_name")

display(df.sample(10))
```

## 3. Filtrando valores en la tabla de datos

Podemos aplicar distintos filtros a la tabla de datos para obtener una nueva tabla específicamente con los valores que nos puedan interesar. Entre estos filtros existen agrupamientos, condicionales, máximos, mínimos, entre otros. 

### Ejemplo 1

Pongamos que quieres solo los datos del mes de **Enero** (de cualquier año) entre las **08::00:00** y las **15:00:00** (¿Por qué?... no sé, solo es un ejemplo XD). A continuacíon ves como hacerlo.  

```python
new_df = df[df["month"] == "January"].between_time('08:00:00','15:00:00')
# Esto se lee: Dame los datos cuyos valores en la columna "month_name" son "January" entre las horas 08:00:00 y 15:00:00
display(new_df)
```

En la tabla anterior puedes ver que ya solo se muestran los datos que comparten el mes de **Enero** dentro del rango de horas requerido.

### Ejemplo 2

Supongamos que de la nueva tabla de datos que generamos anteriormente, quieres eliminar todos aquellas filas que tengan "*default*" en la columna *min_occ_1*. Solo necesitas usar el elemento "inverso" `~` como sigue: 

```python
new_df = df[~df["min_occ_1"].str.contains("default")]
display(new_df)
```

Si vas hasta el final de las tablas, te darás cuenta que el "shape" de la tabla va cambiando (i.e. El número de filas y/o columnas se va reduciendo mientras más filtros pones).

## Ejemplo 3

Finalmente, digamos que quieres agregar los valores **mímimo**, **máximo** y la **media**  de algunas de las columnas y agruparlos por día (en la columna *date*). Ejemplo:

```python
# Expresiones matemáticas a implementar
ops = ["min", "max", "mean"]
new_df_2 = df.groupby("date").agg({"speed": ops, "q_fl_14200_60": ops, "v_fl_14200_60": ops})
display(new_df_2)
```

Este es un ejemplo práctico para que puedas ver cómo hacer distintas cosas con las tablas de Pandas de manera rápida.

## ¿Y para qué todo esto? - ¡Para graficar!

Ya que tenemos la tabla pequeña que está arriba, vamos a hacer distintos tipos de gráficas. Para ello, necesitas importar una librería llamada `matplotlib.pyplot`.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
```

Esta gráfica la haremos únicamente con los datos que corresponden a la columna *speed*. Como ves, *speed* tiene a su vex 3 columnas (*min*, *max* y *mean*).

```python
display(new_df_2["speed"])
```

Una vez que vemos los datos en la tabla, podemos darnos una idea de qué es lo que podemos esperar de la gráfica. La siguiente gráfica será simple: *Gráfica de línea*

```python
# .plot() grafica por default con gráfica de lineas. figsize da el tamaño en la dirección "x" y "y"
grafica = new_df_2["speed"].plot(figsize=(10,8))

# grafica es un objeto de matplotlib el cual contiene diferentes funciones como las siguientes
grafica.set_title('Estadísticas de la velocidad',fontsize= 20)

# Dando nombre al eje Y (Puedes hacer lo mismo para el eje x con "set_xlabel()". En caso de no hacer esto,
# el valor por default es lo que está en index - en este caso "date".)
grafica.set_xlabel("Date", fontsize= 15)
grafica.set_ylabel("Speed", fontsize= 15)

# Para poner los marcadores en el eje x con las fechas (que ahora están en los índices de la tabla)
grafica.set_xticklabels(new_df_2.index.values)
```

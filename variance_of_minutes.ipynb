```python
import pandas as pd
from datetime import datetime
```

# Leyendo la tabla de datos 
1. Leemos la tabla CSV generada con el código de Python y generamos un "Pandas Data frame". Para esto, debemos pasar el nombre del archivo a la función `pandas.read_csv(<nombre del archivo>, sep = <separador>)`.
2. La función `display()` muestra la tabla con un formato reconocido por el Jupyter notebook.

```python
# Comenzamos poniendo el nombre del archivo csv que queremos leer. Si no se encuentra en la misma carpeta que
# este notebook, entonces debes poner el path global del archivo
file = "~/Downloads/lanes_info_original_noholidays.csv"

df = pd.read_csv(file, sep=',') # Recuerda, se necesita especificar cual es el separador de columnas

display(df)
```

Ahora, tomaremos únicamente las columnas que nos interezan para calcular la varianza. Posteriormente, volveremos a unir las columnas resultantes con nuestra tabla original.

```python
# Dejamos únicamente las columnas que nos interesan momentaneamente
df2 = df[["time","travel_time_hw"]]
display(df2)
```

En esta tabla te pongo la *varianza* (y la *deviación estandar* en caso que la necesites)

```python
# var: variance
# std: standard deviation
#
### Los valores que ves en la tabla df3 son los de la varianza y la desviación estandar de los 1440 minutos del día
df3 = df.groupby(df["time"]).agg({"travel_time_hw":["var", "std"]})
display(df3)

# De la siguiente manera, se logra que la varianza y la desviación estandar se calcule para cada minuto durante el
# año, pero de igual forma se repita cada que se encuentre el mismo minuto
df4 = df2.set_index("time").groupby(level="time").transform("var").rename(columns={"travel_time_hw":"travel_time_hw_var"})
df4 = df4.reset_index()
display(df4)

df5 = df2.set_index("time").groupby(level="time").transform("std").rename(columns={"travel_time_hw":"travel_time_hw_std"})
df5 = df5.reset_index()
display(df5)
```

Finalmente, unimos las dos nuevas columnas (cada una se encuentra en un data frame distinto) a nuestra tabla global.

```python
# Dado que los índices son los mismos en la tabla original como en las dos nuevas tablas
# podemos declarar left_index=True y right_index=True. Esto significa que la función merge
# va a hacer una relación uno a uno entre los índices de las tablas y unirá (ej. la columna
# travel_time_hv_var) la tabla ej. df4 en df.
df = df.merge(df4["travel_time_hw_var"], left_index=True, right_index=True)
df = df.merge(df5["travel_time_hw_std"], left_index=True, right_index=True)

# Ordenamos las columnas nuevamente (ya que las nuevas columnas se agregaron al final)
final_result = df.reindex(sorted(df.columns), axis=1)

display(final_result)
```

Corre la siguiente celda para guardar la tabla en formato CSV. Se descargará en la misma ubicación donde tengas este Jupyter-notebook

```python
final_result.to_csv("Tabla_final_final.csv", sep=';')
```
